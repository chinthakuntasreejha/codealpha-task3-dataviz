```python
# üìå TASK 3: DATA VISUALIZATION
# Using Titanic Dataset (Online)

# Step 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Optional for beautiful visuals
sns.set(style="whitegrid", palette="Set2")
plt.rcParams['figure.figsize'] = (10, 6)

# Step 2: Load Dataset
url = "https://calmcode.io/static/data/titanic.csv"
df = pd.read_csv(url)

# Step 3: Basic Info (Optional)
print("Shape:", df.shape)
print(df.info())
print(df.describe())
print("\nMissing Values:\n", df.isnull().sum())

# Step 4: Data Visualizations

# 1Ô∏è‚É£ Survival by Gender
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Passenger Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

# 2Ô∏è‚É£ Age Distribution with KDE
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Passenger Count")
plt.show()

# 3Ô∏è‚É£ Passenger Class vs Survival
sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Passenger Class vs Survival")
plt.xlabel("Class (1 = Upper, 3 = Lower)")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

# 4Ô∏è‚É£ Boxplot: Age by Class
sns.boxplot(x='pclass', y='age', data=df)
plt.title("Age Distribution by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()

# 5Ô∏è‚É£ Fare Distribution
sns.histplot(df['fare'], bins=40, kde=True, color='purple')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()

# 6Ô∏è‚É£ Correlation Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# 7Ô∏è‚É£ Pie Chart: Class Distribution
labels = df['pclass'].value_counts().index
sizes = df['pclass'].value_counts().values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title("Passenger Class Distribution")
plt.axis('equal')
plt.show()

# Step 5: Save the final CSV (optional)
df.to_csv("titanic_visualized.csv", index=False)
```
